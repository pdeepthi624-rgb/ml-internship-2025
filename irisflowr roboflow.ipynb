```python
import os
import cv2
import matplotlib.pyplot as plt
from roboflow import Roboflow

# Connect to Roboflow and download dataset
rf = Roboflow(api_key="Wj826uI6dgjGzFdLZGBn")
project = rf.workspace("developer-3yo81").project("iris-flower-oxdjw")
version = project.version(1)
dataset = version.download("coco")  # Downloads dataset in COCO format

# ✅ Correct path (with 'images')
image_dir = os.path.join(dataset.location, "train", "images")

# Check if directory exists
if not os.path.exists(image_dir):
    print(f"Directory not found: {image_dir}")
else:
    # List all images
    image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

    if not image_files:
        print("No images found in directory.")
    else:
        # Show first 5 images
        for i in range(min(5, len(image_files))):
            img_path = os.path.join(image_dir, image_files[i])
            img = cv2.imread(img_path)

            if img is None:
                print(f"Failed to read image: {img_path}")
                continue

            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            plt.figure(figsize=(6, 6))
            plt.imshow(img)
            plt.title(f"Image: {image_files[i]}")
            plt.axis("off")
            plt.show()
```

```python
# Install Roboflow
!pip install roboflow

# Import necessary modules
from roboflow import Roboflow
import matplotlib.pyplot as plt
import cv2
import os

# Connect to Roboflow and download dataset
rf = Roboflow(api_key="Wj826uI6dgjGzFdLZGBn")
project = rf.workspace("developer-3yo81").project("iris-flower-tykcj")
version = project.version(1)
dataset = version.download("coco")  # This downloads the dataset

# 🔍 Smart function to locate the image folder
def find_image_dir(root_path):
    for dirpath, dirnames, filenames in os.walk(root_path):
        if any(fname.lower().endswith(('.jpg', '.jpeg', '.png')) for fname in filenames):
            return dirpath
    return None

# Try to find training images
image_dir = find_image_dir(dataset.location)

if image_dir is None:
    print("❌ No image folder found in dataset.")
else:
    print(f"✅ Found image folder: {image_dir}")

    # List all image files
    image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

    if not image_files:
        print("❌ No images found in the folder.")
    else:
        # Display first 5 images
        for i in range(min(5, len(image_files))):
            img_path = os.path.join(image_dir, image_files[i])
            img = cv2.imread(img_path)

            if img is None:
                print(f"⚠️ Failed to read image: {img_path}")
                continue

            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            plt.figure(figsize=(6, 6))
            plt.imshow(img)
            plt.title(f"Image: {image_files[i]}")
            plt.axis("off")
            plt.show()
```
